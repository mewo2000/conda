```python
input=open("input_03.txt", "r")


def split(string):
    return[char for char in string]

il=[]
for line in input:
    value=line.strip()
    #value= split(value)
    il.append(value)
#print(il)

ox = []
co = []

#list of all the bits in position of given integer
def sbits(l,i): 
    a=[]
    s=[]
    for item in l:
        a = split(str(item))
        s.append(int(a[i]))
    return(s)
        
#determines the most common bit in given list
def common(l):
    x = sum(l)
    if (x==500):
        return -1
    if (x>500):
        return 1
    if (x<500):
        return 0

def cleaned(string):
    a = []
    a = split(string)
    #print(a)
    return a
    
ox = il.copy()
#print(ox)
#print(len(ox))
j=0
while(len(ox)!=1):
    #print(len(ox))
    for k in ox:
        #print(ox.index(k))
        a = cleaned(k)
        #print(j)
        if(common(sbits(ox,j))== -1):
            if(int(a[j]) == 0):
                ox.pop(ox.index(k))
        else:
            if(int(a[j]) != common(sbits(ox,j))):
                if(j <= 12):
                    ox.pop(ox.index(k))
    #print(len(ox))
    j=j+1
    if(j>=12):
        print(ox)
        break
#print(ox)
    

```
