```python
input = open("input_03.txt", "r")

il = []

for line in input:
    value = line.strip()
    il.append(value)
    
input.close
#print(il)

def split(string):
    return[char for char in string]

def convert(s):
  
    # initialization of string to ""
    new = ""
  
    # traverse in the string 
    for x in s:
        new += x 
  
    # return string 
    return new

a = []
gamma = []
epsilon = []
x = 0

for i in range(12):
    x = 0
    for item in il:
        a = split(str(item))
        x = x + int(a[i])
    if x > 500:
        g=1
        gamma.append(g)
        e=0
        epsilon.append(e)
    else:
        g=0
        gamma.append(g)
        e=1
        epsilon.append(e)
        
        
string_ints_gamma = [str(int) for int in gamma]
gamma_f = int(convert(string_ints_gamma), 2)
#print("gamma  ", gamma_f)

string_ints_ep = [str(int) for int in epsilon]
ep_f = int(convert(string_ints_ep), 2)
#print("epsilon", ep_f)

print(ep_f*gamma_f)
```
